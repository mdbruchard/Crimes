# Context

`DR_NO`: Division of Records Number: Official file number made up of a 2 digit year, area ID, and 5 digits

`Date Rptd`: MM/DD/YYYY

`DATE OCC`: MM/DD/YYYY

`TIME OCC`: In 24 hour military time.

`AREA`: The LAPD has 21 Community Police Stations referred to as Geographic Areas within the department. These Geographic Areas are sequentially numbered from 1-21.

`AREA NAME`: The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. For example 77th Street Division is located at the intersection of South Broadway and 77th Street, serving neighborhoods in South Los Angeles.

`Rpt Dist No`: A four-digit code that represents a sub-area within a Geographic Area. All crime records reference the "RD" that it occurred in for statistical comparisons. Find LAPD Reporting Districts on the LA City GeoHub at http://geohub.lacity.org/datasets/c4f83909b81d4786aa8ba8a74a4b4db1_4

`Part 1-2`:	

`Crm Cd`: Indicates the crime committed. (Same as Crime Code 1)

`Crm Cd Desc`: Defines the Crime Code provided.

`Mocodes`: Modus Operandi: Activities associated with the suspect in commission of the crime.See attached PDF for list of MO Codes in numerical order. https://data.lacity.org/api/views/y8tr-7khq/files/3a967fbd-f210-4857-bc52-60230efe256c?download=true&filename=MO%20CODES%20(numerical%20order).pdf

`Vict Age`: Two character numeric

`Vict Sex`: F - Female M - Male X - Unknown

`Vict Descent`: Descent Code: A - Other Asian B - Black C - Chinese D - Cambodian F - Filipino G - Guamanian H - Hispanic/Latin/Mexican I - American Indian/Alaskan Native J - Japanese K - Korean L - Laotian O - Other P - Pacific Islander S - Samoan U - Hawaiian V - Vietnamese W - White X - Unknown Z - Asian Indian

`Premis Cd`: The type of structure, vehicle, or location where the crime took place.

`Premis Desc`: Defines the Premise Code provided.

`Weapon Used Cd`: The type of weapon used in the crime.

`Weapon Desc`: Defines the Weapon Used Code provided.

`Status Desc`: Defines the Status Code provided.

`Crm Cd 1`: Indicates the crime committed. Crime Code 1 is the primary and most serious one. Crime Code 2, 3, and 4 are respectively less serious offenses. Lower crime class numbers are more serious.

`Crm Cd 2`: May contain a code for an additional crime, less serious than Crime Code 1.

`Crm Cd 3`: May contain a code for an additional crime, less serious than Crime Code 1.

`Crm Cd 4`: May contain a code for an additional crime, less serious than Crime Code 1.

`LOCATION`: Street address of crime incident rounded to the nearest hundred block to maintain anonymity.

`Cross Street`: Cross Street of rounded Address

`LAT`: Latitude

`LON`: Longtitude

# Fixing inconsistence/Dtype

```python
data['Date Rptd'] = pd.to_datetime(data['Date Rptd'])
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'])
```

```python
data.head()
```

```python
# Fuction to extract only the hour OCC
def time(time):
    # Turn the numbers into string
    time = str(time)
    
    # If the first to indexes are higher than 24 return 
    if int(time[:2]) >= 24:
        return int(time[0]) # Return only the first index
    else:
        return int(time[:2]) # Else return the two first indexes
```

```python
data['TIME OCC'] = data['TIME OCC'].apply(time)
data.sample(10)
```

```python
# Eliminate the space between the column names
for col in data.columns:
    data.rename(columns={col: col.replace(' ', '_')}, inplace=True)
```

```python
data.columns
```

```python
# Difference of days between the occurence and the report
data['days_occ_to_rptd'] = data['Date_Rptd'] - data['DATE_OCC']  
```

# Analysis &  Visualization

## Period with more crimes

```python
# Grouping dates
date_crime = data.groupby(by='DATE_OCC')['Date_Rptd'].count().reset_index()

# Saving each year in a variable
crime_2020 = date_crime[date_crime['DATE_OCC'].dt.year == 2020]
crime_2021 = date_crime[date_crime['DATE_OCC'].dt.year == 2021]
crime_2022 = date_crime[date_crime['DATE_OCC'].dt.year == 2022]

# Plotting crime over the years
plt.figure(figsize=(14, 6))


sns.lineplot(data=crime_2020, x='DATE_OCC', y='Date_Rptd')

plt.xticks(rotation=45)


sns.lineplot(data=crime_2021, x='DATE_OCC', y='Date_Rptd')

plt.xticks(rotation=45)


sns.lineplot(data=crime_2022, x='DATE_OCC', y='Date_Rptd')
plt.title('Crimes Over the Years')
plt.xticks(rotation=45);
```

Sightly increase the number of crime over the years. At the end of 2022 getting the record in this data of over $1100$ cases.

Notice, in around April 2020 there a decrease of criminality maybe caused by Covid

## Time the crime normally happen

```python
sns.countplot(data=data, x='TIME_OCC', palette='tab10')
plt.title('Number of crime per time')
plt.xlabel('Time in Hours')
plt.ylabel('Number of Occurences');
```

Most of them occur during the day when there are more people on the streets. The time with higher number of crimes is between 12pm and 13pm followed by 20pm and 21pm

The mean is around $40$ and the mode is $30$ in every year in the dataset

# Time from crime occur to report

How long it takes to report the crime?

What kind of crimes take longer to be reported?

```python
data.head()
```

```python
long_period = data.query('days_occ_to_rptd.dt.days > 100')
```

```python
# Join Columns in one text
text = ''.join(long_period['Crm_Cd_Desc'])

# Generate the wordcloud
wc = WordCloud(width=600, height=400, background_color='white', max_words=30).generate(text)

# Plotting the wordcloud
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis('off');
```

Identity theft is the majority of the cases of delay reports. Also, there are crime against children a clearly sexual crimes as well.

```python
sexual_crimes = long_period[long_period['Crm_Cd_Desc'].str.contains('SEX')]
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=sexual_crimes, x='Vict_Sex')
plt.xlabel('Sex')
plt.title('Sex Counting')

sexual_crimes_filtered = sexual_crimes[sexual_crimes['Vict_Age'].between(1, 100)]

plt.subplot(1, 2, 2)
sns.histplot(data=sexual_crimes_filtered, x='Vict_Age', bins=30)
plt.title('Victims Age')
plt.xlabel('Age');
print(f'The average of victims age is {round(sexual_crimes_filtered.Vict_Age.mean(), 2)}')
```

```python
print(f'The average of days until those case get reported is {sexual_crimes_filtered.days_occ_to_rptd.mean()}')
```
